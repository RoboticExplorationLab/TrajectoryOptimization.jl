```julia
using TrajectoryOptimization
using Plots
```

```julia
n = 2 # number of pendulum states
m = 1 # number of pendulum controls
model! = Model(Dynamics.pendulum_dynamics!,n,m) # inplace dynamics model

opts = SolverOptions()
opts.square_root = false
opts.verbose = true
opts.cache = true
opts.cache=true
opts.c1=1e-4
opts.c2=2.0
opts.mu_al_update = 100.0
opts.infeasible_regularization = mean(Dynamics.pendulum[2].R)
opts.eps_constraint = 1e-3
opts.eps = 1e-5
opts.iterations_outerloop = 250
opts.iterations = 1000

# Constraints
u_min = -10
u_max = 10
x_min = [-100;-100]
x_max = [100; 100]
obj_uncon = Dynamics.pendulum[2]

obj_uncon.R[:] = [1e-3]

obj = ConstrainedObjective(obj_uncon, u_min=u_min, u_max=u_max, x_min=x_min, x_max=x_max)

solver = Solver(model!,obj,dt=0.1,opts=opts)

U = ones(m,solver.N-1)
X = ones(n,solver.N)

X_interp = line_trajectory(solver.obj.x0,solver.obj.xf,solver.N)
@time results = solve_al(solver,X_interp,U)
```

```julia
plot(results.X',title="Final Solve (not properly regularized control)",ylabel="x(t)")
```

```julia
plot(results.U',title="Final Solve (not properly regularized control)",ylabel="u(t)")
```

```julia
println(results.X[:,end])
idx = find(x->x==2,results.iter_type)
# plot(results.result[end].X')

norm(results.result[idx[1]].U-results.result[end].U)

plot(results.result[idx[1]].U',label="Control Infeasible Solve",color="green")
plot!(results.result[end].U',label="Control Dynamically Constrained",color="red")
```

```julia
tmp = ConstrainedResults(solver.model.n,solver.model.m,size(results.result[1].C,1),solver.N)
tmp.U[:,:] = results.result[idx[1]].U
tmp2 = ConstrainedResults(solver.model.n,solver.model.m,size(results.result[1].C,1),solver.N)
tmp2.U[:,:] = results.result[end].U

rollout!(tmp,solver)
plot(tmp.X',title="Rollouts-bad regularization",label="Infeasible rollout")

rollout!(tmp2,solver)
plot!(tmp2.X',label="Dynamically constrained rollout")
```

Proper Regularization on Control

```julia
obj_uncon.R[:] = [1e-2]
obj = ConstrainedObjective(obj_uncon, u_min=u_min, u_max=u_max, x_min=x_min, x_max=x_max)

solver = Solver(model!,obj,dt=0.1,opts=opts)
@time results = solve_al(solver,X_interp,U)
```

```julia
plot(results.X',title="Final Solve (properly regularized control)",ylabel="x(t)")
```

```julia
plot(results.U',title="Final Solve (properly regularized control)",ylabel="u(t)")
```

```julia
println(results.X[:,end])
idx = find(x->x==2,results.iter_type)
# plot(results.result[end].X')

norm(results.result[idx[1]].U-results.result[end].U)

plot(results.result[idx[1]].U',label="Control Infeasible Solve",color="green")
plot!(results.result[end].U',label="Control Dynamically Constrained",color="red")
```

```julia
tmp = ConstrainedResults(solver.model.n,solver.model.m,size(results.result[1].C,1),solver.N)
tmp.U[:,:] = results.result[idx[1]].U
tmp2 = ConstrainedResults(solver.model.n,solver.model.m,size(results.result[1].C,1),solver.N)
tmp2.U[:,:] = results.result[end].U

rollout!(tmp,solver)
plot(tmp.X',title="Rollouts- good regularization",label="Infeasible rollout")

rollout!(tmp2,solver)
plot!(tmp2.X',label="Dynamically constrained rollout")
```
